```python
import json
json_eles = []
json_lines = []
inputfilepath = '/home/preslav/Biotokens_1/ChineseNMT/data/json/test.txt'
outfilepath = '/home/preslav/Biotokens_1/ChineseNMT/data/json/mytest.json'
f = open(inputfilepath, 'r')
lines = f.readlines()
print(lines)

for i in lines:
    json_eles = []
    json_eles.append(i.split("\t")[1].split('\n')[0])
    json_eles.append(i.split("\t")[0])
    json_lines.append(json_eles)

print(json_lines)

with open(outfilepath, 'w') as output_file:
	json.dump(json_lines, output_file, indent=2)
```
